#### Take mean of last 25 data points (That is the correct location of the landmark)

```python
lm00_xy = [np.mean(lm00_x[-25:]),np.mean(lm00_y[-25:])]
lm01_xy = [np.mean(lm01_x[-25:]),np.mean(lm01_y[-25:])]
lm02_xy = [np.mean(lm02_x[-25:]),np.mean(lm02_y[-25:])]
lm03_xy = [np.mean(lm03_x[-25:]),np.mean(lm03_y[-25:])]
lm04_xy = [np.mean(lm04_x[-25:]),np.mean(lm04_y[-25:])]
lm05_xy = [np.mean(lm05_x[-25:]),np.mean(lm05_y[-25:])]
lm06_xy = [np.mean(lm06_x[-25:]),np.mean(lm06_y[-25:])]
lm07_xy = [np.mean(lm07_x[-25:]),np.mean(lm07_y[-25:])]
lm08_xy = [np.mean(lm08_x[-25:]),np.mean(lm08_y[-25:])]
lm09_xy = [np.mean(lm09_x[-25:]),np.mean(lm09_y[-25:])]
lm10_xy = [np.mean(lm10_x[-25:]),np.mean(lm10_y[-25:])]
lm11_xy = [np.mean(lm11_x[-25:]),np.mean(lm11_y[-25:])]
```

#### Extract all landmark location 

```python
all_lm_x = [lm00_xy[0],lm01_xy[0],lm02_xy[0],lm03_xy[0],lm04_xy[0],lm05_xy[0],
            lm06_xy[0],lm07_xy[0],lm08_xy[0],lm09_xy[0],lm10_xy[0],lm11_xy[0]]

all_lm_y = [lm00_xy[1],lm01_xy[1],lm02_xy[1],lm03_xy[1],lm04_xy[1],lm05_xy[1],
            lm06_xy[1],lm07_xy[1],lm08_xy[1],lm09_xy[1],lm10_xy[1],lm11_xy[1]]
```

```python
plt.plot(lm00_xy[0],lm00_xy[1],'X',markersize=8,color='g',label='Start')
plt.plot(all_lm_x[1:-1],all_lm_y[1:-1],'X',markersize=8,color='r')
plt.plot(lm11_xy[0],lm11_xy[1],'X',markersize=8,color='b',label='end')
plt.legend()
plt.xlabel('Landmark x coordinates (m)')
plt.ylabel('Landmark y coordinates (m)')
plt.grid()
```

#### Create a dataframe and store it as csv

```python
radar_xy = pd.DataFrame(columns=['x','y'])
radar_xy['x'] = all_lm_x
radar_xy['y'] = all_lm_y
radar_xy
```
