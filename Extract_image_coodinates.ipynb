```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO
import pandas as pd
```

#### Import YOLO-v8

```python
# Load a model
model = YOLO("yolov8n.yaml")  # build a new model from scratch
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

# Use the model
model.train(data="coco128.yaml", epochs=3)  # train the model
metrics = model.val()  # evaluate model performance on the validation set
results = model("https://ultralytics.com/images/bus.jpg")  # predict on an image
# path = model.export(format="onnx")  # export the model to ONNX format
```

### Define image datapaths

```python
image_data_path = ["Raspi_data/2023-12-10-14-55-35.jpg",
                   "Raspi_data/2023-12-10-14-55-44.jpg",
                   "Raspi_data/2023-12-10-14-55-53.jpg",
                   "Raspi_data/2023-12-10-14-56-02.jpg",
                   "Raspi_data/2023-12-10-14-56-11.jpg",
                   "Raspi_data/2023-12-10-14-56-19.jpg",
                   "Raspi_data/2023-12-10-14-56-28.jpg",
                   "Raspi_data/2023-12-10-14-56-37.jpg",
                   "Raspi_data/2023-12-10-14-56-46.jpg",
                   "Raspi_data/2023-12-10-14-56-55.jpg",
                   "Raspi_data/2023-12-10-14-57-03.jpg",
                   "Raspi_data/2023-12-10-14-57-12.jpg"]
```

#### Extract bounding boxes

```python
image_x = []
image_y = []
```

```python
for image_path in image_data_path:
    results = model(image_path)
    boxes = results[0].boxes
    for box in boxes:
        if box.cls.item()==0: ## 0 is class id for person
            print(box.xywh.cpu().detach().numpy())
            cc = box[0].xywh.cpu().detach().numpy()[0]
            image_x.append(cc[0])
            image_y.append(cc[1])
            break ## To exit from the for loop if a single person is detected twice

    ## Show the deected bounding boxes
    plt.figure()
    plt.imshow(cv2.cvtColor(results[0].plot(),cv2.COLOR_BGR2RGB))
    plt.xticks([]);plt.yticks([])
```

#### Create a dataframe and store it as csv

```python
image_xy = pd.DataFrame(columns=['x','y'])
image_xy['x'] = image_x
image_xy['y'] = image_y
image_xy
```

```python
image_xy.to_csv('image_xy.csv',index=False)
```
